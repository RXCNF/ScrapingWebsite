## **TEXT SCRAPING WEBSITE BRILIO FOOD - USING HTML**


> Muhammad Raihan Fasya - 23031554037



Mengimpor Library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Fungsi untuk scraping satu halaman

```python
def scrape_single_page(page_number):
    url = f"https://www.briliofood.net/resep/index{page_number}.html"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        recipes = soup.find_all('div', class_='item__description')

        scraped_data = []
        for recipe in recipes:
            a_tag = recipe.find('a')
            date_tag = recipe.find('span')

            if a_tag:
                title = a_tag.get_text(strip=True)
                link = a_tag['href']
                upload_date = date_tag.get_text(strip=True) if date_tag else "Tanggal tidak tersedia"
                scraped_data.append({
                    'Judul': title,
                    'Link': link,
                    'Halaman': page_number,
                    'Tanggal Upload': upload_date
                })

        return scraped_data
    else:
        print(f"Gagal mengakses halaman {page_number}.")
        return []
```

Fungsi untuk menyimpan ke CSV menggunakan pandas

```python
def save_to_csv(data):
    df = pd.DataFrame(data)
    df.to_csv('scraped_recipes.csv', index=False, encoding='utf-8')
```

Fungsi utama untuk mengambil input pengguna

```python
def main():
    try:
        start_page = int(input("Masukkan nomor halaman awal yang ingin Anda scraping dari web briliofood: "))
        end_page = int(input("Masukkan nomor halaman akhir yang ingin Anda scraping dari web briliofood: "))

        all_data = []
        for page in range(start_page, end_page + 1):
            all_data.extend(scrape_single_page(page))

        save_to_csv(all_data)

        df = pd.DataFrame(all_data)
        display(df)
        print("Scraping selesai, data disimpan di scraped_recipes.csv.")

    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")
```

Menjalankan Program

```python
if __name__ == "__main__":
    main()
```
